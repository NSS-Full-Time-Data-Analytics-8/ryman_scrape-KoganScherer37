```python
#pull headliners code
headliners = soup.findAll('a',attrs = {'class':'tribe-event-url'})
headliners
```

```python
#Bands in headliners
bands = [bands.get('title') for bands in headliners]
bands
```

```python
#pull time code
time = soup.findAll('time')
time
```

```python
#date and time in time
date_time = [date_time.get('datetime') for date_time in time]
date_time
```

```python
#combine headliners and date and time into dictionary
bands_dict = dict(zip(bands,date_time))
bands_dict
```

```python
#turning previous dictinary made into a dataframe
bands_df = pd.DataFrame.from_dict(bands_dict, orient = 'index')
bands_df.columns = ['date_and_time']
bands_df
```

```python
#reset df index
bands_df = bands_df.reset_index()
bands_df
```

```python
#clean up column names
bands_df = bands_df.rename(columns = {'index': 'headliners'})
bands_df
```

```python
#split date and time column
bands_df[['date','time']] = bands_df['date_and_time'].str.split(' ',n=1,expand = True)
bands_df
```

```python
#pull preferred columns
bands_df = bands_df.loc[:,['headliners','date','time']]
bands_df
```

```python
#converting date column to datetime
bands_df['date'] = pd.to_datetime(bands_df['date'])
```

```python
#datetime conversion sanity check
bands_df.info()
```

```python
#converting date column date type
bands_df['date'] = bands_df['date'].dt.strftime('%A %B %d, %Y')
bands_df
```

```python
#buckets
headliner = []
date = []
opener = []
```

```python
pages = np.arange(1,5)
```

```python
#creating for loop to pull from multiple pages
for page in pages:
    page = requests.get('https://ryman.com/events/list/?tribe_event_display=list&tribe_paged='+str(page)+'')
    soups = BS(page.text, 'html.parser')
    show_data = soups.find_all('div',attrs = {'class': 'tribe-beside-image'})
    sleep(randint(2,8))
    
#storage for loop
    for store in show_data:
        
        #headliner
        hl_name = store.h2.a.text
        headliner.append(hl_name)
        
        #show date
        show_date = store.find('div', class_ = 'tribe-event-schedule-details').text
        date.append(show_date)
        
        #opener
        show_opener = store.find('span', class_ = 'opener').text if store.find('span', class_ = 'opener') else 'None'
        #print(show_opener)
        opener.append(show_opener)
```

```python
show_data = pd.DataFrame({'headliner': headliner,'date_and_time' : date})
show_data
```

```python
show_data['headliner'] = show_data['headliner'].str.replace('\n','')
show_data
```

```python
show_data['headliner'] = show_data['headliner'].str.replace('\t','')
show_data
```

```python
show_data['date_and_time'] = show_data['date_and_time'].str.replace('\n','')
show_data
```

```python
show_data[['date','time']] = show_data['date_and_time'].str.split(' at ',n=1,expand = True)
show_data
```

```python
show_data = show_data.loc[:,['headliner','date','time']]
show_data
```
